```python
import pandas as pd
import plotly.express as px
```

```python
df.info()
df['Published Date'] = pd.to_datetime(df['Published Date'])
df['Year'] = df['Published Date'].dt.year
df['Month'] = df['Published Date'].dt.month
df['LikeRatio'] = df['Likes'] / df['Views']
df['Engagement'] = (df['Likes'] + df['Comments']) / df['Views']
```

# Yillar bo‚Äòyicha trend

```python
import pandas as pd
import plotly.express as px

# Sana ustunini datetime formatiga o'tkazamiz
df['Published Date'] = pd.to_datetime(df['Published Date'], errors='coerce')

# Yilni ajratib olish
df['Year'] = df['Published Date'].dt.year

# Har bir yil bo‚Äòyicha ko‚Äòrishlar summasi
yearly_views = df.groupby('Year', as_index=False)['Views'].sum()

# Eng ko‚Äòp ko‚Äòrilgan yilni aniqlash
top_year = yearly_views.loc[yearly_views['Views'].idxmax()]
max_year = int(top_year['Year'])
max_views = int(top_year['Views'])

print(f"Eng ko‚Äòp ko‚Äòrilgan yil: {max_year} ‚Äî {max_views:,} ta ko‚Äòrish")

# Grafik chizish (Bar chart)
fig = px.bar(
    yearly_views,
    x='Year',
    y='Views',
    text='Views',
    title='üìä Yillar Bo‚Äòyicha Umumiy Ko‚Äòrishlar (Eng Ko‚Äòp Ko‚Äòrilgan Yil Ajratilgan)',
    color='Views',
    color_continuous_scale='Blues'
)

# Faqat bar‚Äôlar uchun text joylashuvini sozlash
fig.update_traces(texttemplate='%{text:,}', textposition='outside')

# Eng ko‚Äòp bo‚Äòlgan yilni ajratib ko‚Äòrsatish (scatter marker bilan)
fig.add_scatter(
    x=[max_year],
    y=[max_views],
    mode='markers+text',
    text=[f"üî• {max_year}\n{max_views:,} ko‚Äòrish"],
    textposition='top center',
    marker=dict(size=15, color='red', symbol='star'),
    name='Eng ko‚Äòp ko‚Äòrilgan yil'
)

# Grafik dizayni
fig.update_layout(
    xaxis_title='Yil',
    yaxis_title='Ko‚Äòrishlar soni',
    plot_bgcolor='rgba(250,250,250,0.9)',
    title_font=dict(size=22, color='darkblue'),
    font=dict(size=14),
    showlegend=False
)

fig.show()
```

# Top Views, Likes va Comments

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# üîπ 1. Top 10 eng ko‚Äòp ko‚Äòrilgan videolar
top_views = df.nlargest(10, 'Views')

# üîπ 2. Top 10 eng ko‚Äòp like olgan videolar
top_likes = df.nlargest(10, 'Likes')

# üîπ 3. Top 10 eng ko‚Äòp izoh olgan videolar
top_comments = df.nlargest(10, 'Comments')

# üî∏ Subplot (3 ta ustunli) hosil qilamiz
fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=("üëÅÔ∏è Eng ko‚Äòp KO‚ÄòRILGAN videolar", 
                    "‚ù§Ô∏è Eng ko‚Äòp LIKE olgan videolar", 
                    "üí¨ Eng ko‚Äòp IZOH olgan videolar"),
    shared_yaxes=False
)

# 1Ô∏è‚É£ Views
fig.add_trace(
    go.Bar(
        x=top_views['Views'],
        y=top_views['Video Title'],
        orientation='h',
        marker_color='cornflowerblue',
        text=top_views['Views'],
        textposition='outside',
        name='Views'
    ),
    row=1, col=1
)

# 2Ô∏è‚É£ Likes
fig.add_trace(
    go.Bar(
        x=top_likes['Likes'],
        y=top_likes['Video Title'],
        orientation='h',
        marker_color='lightcoral',
        text=top_likes['Likes'],
        textposition='outside',
        name='Likes'
    ),
    row=1, col=2
)

# 3Ô∏è‚É£ Comments
fig.add_trace(
    go.Bar(
        x=top_comments['Comments'],
        y=top_comments['Video Title'],
        orientation='h',
        marker_color='mediumseagreen',
        text=top_comments['Comments'],
        textposition='outside',
        name='Comments'
    ),
    row=1, col=3
)

# üîß Grafikni sozlaymiz
fig.update_layout(
    height=800,
    width=1800,
    title_text="üèÜ YouTube'dagi Top 10 Videolar: Views, Likes va Comments bo‚Äòyicha",
    showlegend=False,
    plot_bgcolor='rgba(250,250,250,0.95)',
    title_font=dict(size=24, color='darkblue')
)

# üî§ O‚Äòqlarni aylantirish
fig.update_yaxes(categoryorder='total ascending')

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# 'Published Date' ni datetime formatga o'tkazamiz
df['Published Date'] = pd.to_datetime(df['Published Date'], errors='coerce')

# Eng eski va eng yangi videolarni topamiz
oldest_video = df.loc[df['Published Date'].idxmin()]
newest_video = df.loc[df['Published Date'].idxmax()]

# Jadval uchun ma'lumotlar
table_data = pd.DataFrame([
    {
        "Turi": "üéûÔ∏è Eng eski video",
        "Sana": oldest_video['Published Date'].date(),
        "Sarlavha": oldest_video['Video Title'],
        "Kanal": oldest_video['Channel'],
        "Ko‚Äòrishlar": oldest_video['Views'],
        "Layklar": oldest_video['Likes'],
        "Izohlar": oldest_video['Comments'],
        "Davlat": oldest_video['Country']
    },
    {
        "Turi": "üÜï Eng yangi video",
        "Sana": newest_video['Published Date'].date(),
        "Sarlavha": newest_video['Video Title'],
        "Kanal": newest_video['Channel'],
        "Ko‚Äòrishlar": newest_video['Views'],
        "Layklar": newest_video['Likes'],
        "Izohlar": newest_video['Comments'],
        "Davlat": newest_video['Country']
    }
])

# Plotly jadval yasash
fig = go.Figure(data=[go.Table(
    header=dict(
        values=list(table_data.columns),
        fill_color='royalblue',
        align='center',
        font=dict(color='white', size=13)
    ),
    cells=dict(
        values=[table_data[col] for col in table_data.columns],
        fill_color=[['#f5f5f5', '#ffffff'] * (len(table_data)//2 + 1)],
        align='center',
        font=dict(color='black', size=12)
    )
)])

fig.update_layout(title='üìä Eng Eski va Eng Yangi Videolar Jadvali', title_font=dict(size=20))
fig.show()
```
